# 1. Import data

```python
#read panda datframe from csv

data = pd.read_csv('all_players.csv')
data.head(5)
```

# 2. Delete duplicate

```python
# function to delete duplicate
def delete_duplicate(df):
    df_ = df.drop_duplicates()
    return df_

data1 =delete_duplicate(data)

data1
```

# 3. Delete column from the list ['Rank', 'Height', 'Weight', 'Alternative positions', 'Age', 'play', 'style', 'url']

```python
# function to delete columns from the list list_to delete = ['Rank', 'Height', 'Weight', 'Alternative positions', 'Age', 'play style', 'url']
def del_list(df, list_to_delete):
    df_ = df.drop(list_to_delete, axis=1)
    return df_

list_to_delete = ['Rank', 'Height', 'Weight', 'Alternative positions', 'Age', 'play style', 'url']
data2 = del_list(data1, list_to_delete)
data2
```

# 4. Transform "Preferred Foot" column to binary 0 and 1

```python
print(data2['Preferred foot'])
```

```python
# function to transform "Preferred Foot" column to binary 0 and 1
def to_binary(df1):
    df1['Preferred foot'] = df1['Preferred foot'].map({'Right': 1, 'Left': 0})
    return df1

data3 = to_binary(data1)
print(data3['Preferred foot'])
```

```python
for x in range (0,101):
  print(data3['Position'][x])
```

# 5. Extrack all_gk

```python
# function to extract datas from data3 where 'Position' column is GK
def extract_gk(df):
    df_ = df[df['Position'] == 'GK']
    return df_

data4 = extract_gk(data3)
data4['Position']
```

```python
data4
```

```python
#check GK attributes
print(data4.columns)
```

```python
# delete all GK attributes
GK_attributes_list = ['GK Diving', 'GK Handling', 'GK Kicking',
       'GK Positioning', 'GK Reflexes']
all_gk = del_list(data4, GK_attributes_list)
all_gk.columns
```

```python
all_gk.head()
```

# 6. Delete missing values from all_gk

```python
# function to count missing values from all_gk
def count_missing_values(df):
    missing_values = df.isnull().sum()
    return missing_values

missing_values = count_missing_values(all_gk)
missing_values
```

```python
# delete columns with missing values
column_with_missing_values = ['Alternative positions', 'play style']
all_gk = del_list(all_gk, column_with_missing_values)
```

```python
# count missing values after
missing_values = count_missing_values(all_gk)
missing_values
```

 # 2. Analysez les mesures de tendance centrale et de dispersion de la variable Weak Foot pour ces deux groupes

```python
# function to extract data from 'wek foot' column where 'Preferred foot' = 1
def extract_weak_foot_R(df):
    df_ = df[df['Preferred foot'] == 1]
    df_ = df_[df_['Weak foot'] != 0]
    return df_

weak_foot_R = extract_weak_foot_R(all_gk)
weak_foot_R['Preferred foot']

describe_column(weak_foot_R, 'Weak foot')

```

```python
# function to extract data from 'wek foot' column where 'Preferred foot' = 0
def extract_weak_foot_L(df):
    df_ = df[df['Preferred foot'] == 0]
    df_ = df_[df_['Weak foot'] != 0]
    return df_

weak_foot_L = extract_weak_foot_L(all_gk)
weak_foot_L['Preferred foot']

describe_column(weak_foot_L, 'Weak foot')
```

# 3. Matrice de correlation

```python
postes = ['ST', 'CM', 'CB']
competences = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF']
for poste in postes:
    print(f"\nMatrice de corrélation pour le poste {poste}:")

    data_poste = data1[data1['Position'] == poste]

    colonnes_correlation = ['OVR'] + competences
    matrice_corr = data_poste[colonnes_correlation].corr()
    print(matrice_corr)
    plt.figure()
    sns.heatmap(matrice_corr, annot=True, cmap='coolwarm')
    plt.title(f'Matrice de corrélation pour {poste}')
    plt.show()

polyvalence = {}
for poste in postes:
    polyvalence[poste] = data1[data1['Position'] == poste][['OVR'] + competences].corr()['OVR'].abs().sum()

poste_plus_polyvalent = max(polyvalence, key=polyvalence.get)
print(f"\nLe poste le plus polyvalent semble être : {poste_plus_polyvalent}")
```
