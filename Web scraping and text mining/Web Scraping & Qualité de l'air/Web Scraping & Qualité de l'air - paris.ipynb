```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify)
```

```python
#function to scrap iqair
def scrap_iqair(page):


    # Pass 'html1' directly to BeautifulSoup'
    soup1 = BeautifulSoup(page.text, 'html.parser')


    # Measurement
    mes = soup1.findAll(class_="measurement-wrapper__value")


    # dictionary of info
    dict_info = {
        'PM2.5' : mes[0].text,
        'PM10' : mes[1].text,
        'O3' : mes[2].text,
        'NO2' : mes[3].text,
    }


    #dayly_air_quality
    dayly_air_quality = soup1.findAll('div',class_='aqi-value-wrapper__container')
    dayly_air_quality_value = dayly_air_quality[0].text[1:3]


    #'aqi-main-pollutant-text'
    main_pollutant = soup1.findAll('div',class_='aqi-overview-footer-container__main-pollutant')


    p_tags = main_pollutant[0].find_all('p', class_='aqi-text__status')
    main_pollutant_name = p_tags[0].text
    # 'aqi-text__value'
    level_main_pollutant = soup1.find_all('p', class_='aqi-text__value') # Call find_all on the soup1 object
    niveau_principaux_polluants = level_main_pollutant[0].text


    # dict info update
    dict_info_update = {
        'nom_ville' : 'Paris',
        'PM2.5' : mes[0].text,
        'PM10' : mes[1].text,
        'Ozone' : mes[2].text,
        'NO2' : mes[3].text,
        'Indice_qualité': dayly_air_quality_value,
        'nom_polluant_principaux':main_pollutant_name,
        'niveau_principaux_polluants':niveau_principaux_polluants
    }


    return dict_info_update

```

```python
data = scrap_iqair(response)

data
```
